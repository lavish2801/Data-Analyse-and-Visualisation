```python
# use matplotlib

import matplotlib.pyplot as plt
```

```python
# show pie chart for Age vs Total disease

xlabel=df["Age"]
ylabel=df["TotalDisease"]

plt.pie(xlabel,ylabel)
plt.show()
```

```python
#Grid Plotting 

xlabel=df["Age"].head(20)
ylabel=df["TotalDisease"].head(20)
plt.plot(xlabel, ylabel,color='black')
plt.grid(color = 'black')
font1 = {'family':'serif','color':'blue','size':20}
plt.title("Age vs Total Disease" ,fontdict = font1)
plt.xlabel("Age")
plt.ylabel("Total Disease")
plt.show()
```
