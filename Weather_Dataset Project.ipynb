 # Data Visualisation Using MatplotLib

```python
#import library

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot 
xpoints =df["Weather Condition"].head(50)
ypoints =df["Temp_C"].head(50)


plt.plot(xpoints, ypoints)
#plt.show()
```

```python
#Plotting Without Line (represt markers)

xpoints =df["Weather Condition"].head(30)
ypoints =df["Temp_C"].head(30)


plt.plot(xpoints, ypoints,'o') 
# instead of 'o', use many marker reference like * ,'.' ,'X' many more
```

```python
# labelling of graph

xpoints =df["Weather Condition"].head(30)
ypoints =df["Temp_C"].head(30)


plt.plot(xpoints, ypoints) 
plt.xlabel("Weather Condition")
plt.ylabel("Temperature")
plt.title("Graph Using Matplotlib")
```

```python
# can Use font and colour also

xpoints =df["Weather Condition"].head(30)
ypoints =df["Temp_C"].head(30)


plt.plot(xpoints, ypoints,'o') 
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title( "Graph Using Matplotlib",fontdict = font1)
plt.xlabel("Weather Condition", fontdict = font2)
plt.ylabel("Temperature", fontdict = font2)
```
